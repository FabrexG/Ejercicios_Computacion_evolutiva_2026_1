```python
class Persona: 
    def __init__(self, nombre: str, edad: int):
        self.nombre = nombre
        self.edad = edad

    @property
    def nombre(self) -> str:
        return self.__nombre
    
    @nombre.setter
    def nombre(self, nombre_nuevo: str) -> None:
        if len(nombre_nuevo) == 0:
            raise ValueError("El nombre no puede estar vacio")
        
        self.__nombre = nombre_nuevo
    
    @property
    def edad(self) -> int:
        return self.__edad
    
    @edad.setter
    def edad(self, edad_nueva) -> None: 
        if edad_nueva < 0 or  edad_nueva > 120:
            raise ValueError("La edad debe de estar entre el rango de (0 y 120)")
        self.__edad = edad_nueva
    
    def __str__(self) -> str:
        return f"Persona: {self.nombre}, {self.edad}"
    
    def __repr__(self) -> str:
        return f"Persona: Nombre = {self.nombre}, edad = {self.edad}"
    
    def presentarse(self) -> None:
        print(f"Hola buenas tardes soy {self.nombre} y tengo {self.edad}")

```

```python
p1: Persona = Persona("Paulina", 24)
print(p1)
print(repr(p1))
p1.presentarse()
```
