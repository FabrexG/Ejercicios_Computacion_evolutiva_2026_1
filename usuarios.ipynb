```python
class Usuario:
    def __init__(self, username: str, email: str) -> str:
        self.username: str = username
        self.email: str = email
    
    @property
    def username(self) -> str:
        return self.__username 
    
    @username.setter
    def username(self, username: str) -> None:
        self.__username = username

    @property 
    def email(self) -> str:
        return self.__email
    
    @email.setter
    def email(self, email: str) -> None:
        self.__email = email
    
    def obtener_perfil(self) -> str: 
        return f"Usuario: [{self.username}] | Email: [{self.email}]"
    
```

```python
class Admin(Usuario):
    def __init__(self, username: str, email: str, nivel_seguridad: int):
        super().__init__(username, email)
        self.nivel_seguridad: int = nivel_seguridad
    
    @property
    def nivel_seguridad(self) -> int:
        return self.__nivel_seguridad
    
    @nivel_seguridad.setter
    def nivel_seguridad(self, nivel_seguridad: int) -> None:
        if (nivel_seguridad < 1 or nivel_seguridad > 5):
            raise ValueError(f"El nivel de seguridad debe de estar entre 1 y 5. Actualemte {nivel_seguridad}")
        
        self.__nivel_seguridad: int = nivel_seguridad
    
    def obtener_perfil(self):
        return f"SISTEMA CRITICO: " + super().obtener_perfil() +  f" | Nivel de seguridad: [{self.nivel_seguridad}]"
    
    
```

```python
admin: Admin = Admin("Paulina", "paulina@gmail.com", 4)
admin.obtener_perfil()
```

```python
try:
    admin: Admin = Admin("Paulina", "paulina@gmail.com", 10)
    admin.obtener_perfil()
except ValueError as e:
    print(e)
```
