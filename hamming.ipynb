```python
def hamming(cadena1: str, cadena2: str) -> int:
    if len(cadena1) != len(cadena2):
        raise ValueError("Error las cadenas deben de tener el mismo tama√±o")
    
    count: int = 0
    for i in range(len(cadena1)):
        if cadena1[i] != cadena2[i]:
            count += 1
    return count
```

```python
c1: str = "1011101"
c2: str = "1001001"
hamming(c1, c2)
```
