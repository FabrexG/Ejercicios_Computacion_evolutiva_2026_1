```python
class MetodoPago:
    def procesar_pago(self, monto: float) -> None:
        print("Validando fondos con la entidad financiera...")
```

```python
class PagoTarjeta(MetodoPago):
    def procesar_pago(self, monto: float) -> None:
        super().procesar_pago(monto)
        print(f"Transacción de ${monto} aprobada por el banco emisor (Visa/Mastercard).")
```

```python
class PagoCripto(MetodoPago):
    def procesar_pago(self, monto: float) -> None:
        super().procesar_pago(monto)
        print(f"Criptomoneda: Confirmando transferencia de ${monto} en la Blockchain... ¡Pago exitoso!")
```

```python
import random
pasarela_pagos: list = [PagoTarjeta(), PagoCripto(), PagoTarjeta(), PagoCripto(), PagoTarjeta(), PagoCripto()]

for cuenta in pasarela_pagos:
    monto = random.randint(100, 5000)
    cuenta.procesar_pago(monto)
```
