Ejercicio 1: Creación de Población y Operador de Mutación
Contexto en Computación Evolutiva:
Un algoritmo evolutivo inicia con una población inicial aleatoria. En este caso, cada individuo tiene 20 genes y cada gen puede tener tres estados posibles (alelos: 0, 1 o 2), lo cual podría representar, por ejemplo, diferentes tipos de proteínas o niveles de una señal. La Mutación es el proceso que altera estos valores para descubrir nuevas soluciones.
Instrucciones:

Crear la Población: Generar una matriz de 10 x 20 (10 individuos, 20 genes) con valores aleatorios enteros entre 0 y 2 inclusive.
Utilizar los type hints adecuados para asegurar que la matriz sea de enteros.
Exploración de Individuos: Extrae y muestra el cromosoma del tercer individuo. Extraer y mostrar los primeros 5 genes de todos los individuos.
Mutación: Extraer un individuo al azar y altera su último gen con un valor del alelo aleatorio permitido entre 0 y 2 inclusive, pero no puede ser el mismo valor que tiene. (un valor que represente una mutación). Imprimir al individuo mutado para verificar el cambio.

```python
import numpy as np
import numpy.typing as npt

poblacion: npt.NDArray[np.long] = np.random.randint(0, 3, size=(10, 20))
```

```python
import random

mutado = random.sample(poblacion.tolist(), 1)
print(mutado)
print(mutado[0][19])
```

```python


gen_mutar = random.randint(0,3)
print(gen_mutar)
print(mutado[0][19])

if gen_mutar == mutado[0][19]:
    while (gen_mutar == mutado[0][19]):
        gen_mutar = random.randint(0,3)
        
        if mutado[0][19] != gen_mutar:
            mutado[0][19] = gen_mutar
            break
else:
    mutado[0][19] = gen_mutar
mutado
```
