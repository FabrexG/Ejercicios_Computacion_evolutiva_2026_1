```python
def dice(cadena1: str, cadena2: str) -> float:

    interseccion: int = 0

    for c1, c2 in zip(cadena1, cadena2):
        if c1 == c2: 
            interseccion += 1
    return (2 * interseccion) / (len(cadena1) + len(cadena2))
        
```

```python
cadena1: str = "1011101"
cadena2: str = "1001001"
```

```python
print(dice(cadena1, cadena2))
```
