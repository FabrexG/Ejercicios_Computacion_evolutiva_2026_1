```python
# Ejercicio 1 — Generacion de vectores aleatorios
# Implementa una funcion:
# def generar_individuo(n):
# que regrese un vector binario aleatorio de longitud n.
# Ejemplo, resultado esperado: [1,0,1,1,0,0,1]
import numpy as np
def generar_individuo(n):
    individuo = []
    for _ in range(0,n):
        numero = np.random.randint(0,2)
        individuo.append(numero)
    return individuo

generar_individuo(7)
```

```python
# Ejercicio 2 — Poblacion inicial
# Implementa la generacion de una poblacion inicial:
# def inicializar_poblacion(tam_poblacion, n):
# Debe devolver una lista de individuos de tamaño n.
# [
#   [?, ?, ?, ?],   # Individuo 1 (tamaño 4)
#   [?, ?, ?, ?],   # Individuo 2 (tamaño 4)
#   [?, ?, ?, ?]    # Individuo 3 (tamaño 4)
# ]
import numpy as np

def inicializar_poblacion(tam_poblacion, n):
    indiviudos = []
    for _ in range(0, tam_poblacion):
        individuo = []
        for _ in range(0, n):
            numero = np.random.normal()
            individuo.append(numero)
        indiviudos.append(individuo)
    return indiviudos

inicializar_poblacion(3,4)
```

```python
# Ejercicio 3 — Espacio de busqueda
# En una representacion binaria de longitud n, cada individuo es un vector de 0s y 1s. Tener
# exactamente k unos significa seleccionar k posiciones del vector para colocar 1.
# Implementa una funcion que calcule cuantos individuos tienen exactamente k unos, solo
# hacer el calculo numerico:
# def calcula_individuos_k_unos(n, k):
# Ejemplo: calcula individuos k unos(5,2) imprime 10
# Los individuos son: 11000, 10100, 10010, 10001, 01100, 01010, 01001, 00110, 00101, 00011

import math

def calcula_individuos_k_unos(n, k):
    #return math.comb(n,k)
    total_combinaciones = math.factorial(n) // (math.factorial(k) * math.factorial(n-k))
    return total_combinaciones

calcula_individuos_k_unos(50,2)

# Si n = 50, ¿es factible evaluar exhaustivamente todas las soluciones? Justifica usando el
# tamano del espacio de busqueda. 
#Respuesta: No, no es factible evaluar exhaustivamente si contamos con k=2 el espacio a evaluar seria de 2^50 lo cual es invalible
```

```python
# Ejercicio 4 — Mutacion de bits
# Implementa:
# def mutacion(individuo, p):
# Cada bit debe cambiar con probabilidad p. Es decir el bit procesado tienen la probabilidad, por
# ejemplo, p = 0.5 de cambiar.

def mutacion(individuo, p):
    individuo_mutado = []
    for bit in individuo:
        probabilidad = np.random.random(1)
        if probabilidad < p:
            individuo_mutado.append(1 - bit)
        else:
            individuo_mutado.append(bit)
    return individuo_mutado

individuo = [1,0,0,1,0,1,0]
print(f"Antes: {individuo}")
print(f"Despues: {mutacion(individuo, 0.1)}")

```
