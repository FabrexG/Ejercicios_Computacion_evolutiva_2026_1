```python
from math import tan, cos, sin

def pedir_2_numeros() -> float: 
    a: float = float(input("Ingresa un el primer numero: "))
    b: float = float(input("Ingresa un el segundo numero: "))
    return a, b

def pedir_numero() -> float: 
    a: float = float(input("Ingresa el numero de radianes para su operacion: "))
    return a
```

```python
def suma(a: float, b:float) -> float:
    return a + b
    

def resta(a: float, b:float) -> float:
    return a - b

def multiplicacion(a: float, b:float) -> float:
    return a * b

def division(a: float, b:float) -> float:
    if b == 0:
        return -1
    return a / b
```

```python
funciones = {
    'suma': suma,
    'resta': resta,
    'multiplicacion': multiplicacion,
    'division': division,
    'tangente': tan,
    'coseno': cos,
    'seno': sin
}

print("Selecciona alguna opcion: ")
for fun in funciones:
    print(fun)


opcion: str = input("Que funcion desea ajecutar?: ").lower()


if opcion not in ('seno', 'coseno', 'tangente'):
    a,b = pedir_2_numeros()
    resultado = funciones[opcion](a,b)
    print(f"El resultado de {opcion} es {resultado}")

else:
    x = pedir_numero()
    resultado = funciones[opcion](x)
    print(f"El resultado de {opcion} es {resultado}")

```
