```python
import re

# Archivo de entrada y salida
input_file = "dataset_llm.json"
output_file = "datos_llm_v.json"

# Leer el contenido original
with open(input_file, "r", encoding="utf-8") as f:
    data = f.read()

# Separar los objetos JSON usando regex
items = re.findall(r"\{.*?\}", data)

# Guardar cada objeto en una línea
with open(output_file, "w", encoding="utf-8") as f:
    for item in items:
        f.write(item + "\n")

print("Conversión completa. Revisa el archivo:", output_file)
```

```python
import pandas as pd

# Cargar los dos archivos JSON
df1 = pd.read_json("dataset_humor_train.json", lines=True)  # primer archivo
df2 = pd.read_json("humor.json", lines=True)   # segundo archivo

# Concatenar ambos DataFrames
df_total = pd.concat([df1, df2], ignore_index=True)

# Reasignar la columna 'id' para que sea consecutiva
df_total["id"] = range(1, len(df_total) + 1)

# Guardar el resultado en un nuevo archivo
df_total.to_json("Dataset_humor_enr.json", orient="records", lines=True, force_ascii=False)

print("Total de registros:", len(df_total))
```
