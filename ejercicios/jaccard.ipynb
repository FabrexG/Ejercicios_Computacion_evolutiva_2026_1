```python
def jaccard(cadena1: str, cadena2: str) -> float:

    interseccion: int = 0
    union: int = 0

    for c1, c2 in zip(cadena1, cadena2):
        if c1 == c2: 
            interseccion += 1
        union += 2
    return interseccion / union
        
```

```python
cadena1: str = "1011101"
cadena2: str = "1001001"
print(jaccard(cadena1, cadena2))
```
