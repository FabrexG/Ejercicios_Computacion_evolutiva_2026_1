```python
class Mutacion_Real:

    def mutacion_uniforme(individuo, probabilidad):
        genes = individuo[0]
        limites = individuo[1]

        for i in range(len(genes)):
            limite_inferior = limites[i][0]
            limite_superior = limites[i][1]
            p = np.random.random()
            if p < probabilidad:
                gen_nuevo = np.random.uniform(limite_inferior, limite_superior)
                genes[i] = gen_nuevo
        
        individuo[0] = genes

        return individuo
    
    def mutacion_no_uniforme(individuo, probabilidad):
        genes = individuo[0]
        limites = individuo[1]

        for i in range(len(genes)):
            
            x_min = limites[i][0]
            x_max = limites[i][1]

            sigma = ((x_max - x_min) / 6)

            print(sigma)

            p = np.random.random()
            if p < probabilidad:
                valor_muestreado = np.random.normal(scale=sigma)
                gen_nuevo = genes[i] + valor_muestreado

                if gen_nuevo > x_max:
                    genes[i] = x_max
                elif gen_nuevo < x_min:
                    genes[i] = x_min
                else:
                    genes[i] = gen_nuevo
        
        individuo[0] = genes

        return individuo
    
    def sigmas_t(current_sigma: list):
        t: float = 1 / np.sqrt(len(current_sigma))
        for i in range(len(current_sigma)):
            nueva_sigma = current_sigma[i] * math.exp(t * np.random.normal())
            current_sigma[i] = nueva_sigma

            if current_sigma[i] < 0.1:
                current_sigma[i] = 0.1 
        
        return current_sigma

    def mutacion_adaptativa(self, individuo, probabilidad, sigma):
        for i in range(len(individuo)):
            print(sigma)
            p = np.random.random()
            if p < probabilidad:
                gen_nuevo = individuo[i] + (sigma[i] * np.random.normal())
                individuo[i] = gen_nuevo
            sigma = self.sigmas_t(sigma)

        return individuo
```

```python
class Cruza_real:

    def cruza_aritmetica_unica(self, padre_1, padre_2, alfa):
        p1 = np.array(padre_1, dtype=float)
        p2 = np.array(padre_2, dtype=float)

        gen = np.random.randint(len(p1))

        hijo1 = p1.copy()
        hijo2 = p2.copy()

        hijo1[gen] = alfa * p1[gen] + (1 - alfa) * p2[gen]
        hijo2[gen] = alfa * p2[gen] + (1 - alfa) * p1[gen]

        return hijo1, hijo2

    def cruza_aritmetica_simple(self, padre_1, padre_2, alfa):
        p1 = np.array(padre_1, dtype=float)
        p2 = np.array(padre_2, dtype=float)

        gen = np.random.randint(len(p1))

        hijo1 = p1.copy()
        hijo2 = p2.copy()

        hijo1[gen:] = alfa * p1[gen:] + (1 - alfa) * p2[gen:]
        hijo2[gen:] = alfa * p2[gen:] + (1 - alfa) * p1[gen:]

        return hijo1, hijo2

    def cruza_aritmetica_completa(self, padre_1, padre_2, alfa):
        p1 = np.array(padre_1, dtype=float)
        p2 = np.array(padre_2, dtype=float)

        hijo1 = alfa * p1 + (1 - alfa) * p2
        hijo2 = alfa * p2 + (1 - alfa) * p1

        return hijo1, hijo2
    
    
```
