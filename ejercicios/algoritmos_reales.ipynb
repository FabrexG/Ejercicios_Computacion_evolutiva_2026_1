```python
import numpy as np

x = [[3.2, -1.5, 0.8], [(0, 5), (-2, 2), (0, 10)]]
```

```python
def mutacion_uniforme(individuo, probabilidad):
    genes = individuo[0]
    limites = individuo[1]

    for i in range(len(genes)):
        limite_inferior = limites[i][0]
        limite_superior = limites[i][1]
        p = np.random.random()
        if p < probabilidad:
            gen_nuevo = np.random.uniform(limite_inferior, limite_superior)
            genes[i] = gen_nuevo
    
    return genes
```

```python
print(mutacion_uniforme(x, 0.5))
```
