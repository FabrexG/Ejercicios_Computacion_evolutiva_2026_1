```python
import numpy as np
import math

x = [[3.2, -1.5, 0.8], [(0, 5), (-2, 2), (0, 10)]]
```

```python
def mutacion_uniforme(individuo, probabilidad):
    genes = individuo[0]
    limites = individuo[1]

    for i in range(len(genes)):
        limite_inferior = limites[i][0]
        limite_superior = limites[i][1]
        p = np.random.random()
        if p < probabilidad:
            gen_nuevo = np.random.uniform(limite_inferior, limite_superior)
            genes[i] = gen_nuevo
    
    individuo[0] = genes

    return individuo
```

```python
print(f"Antes de mutar: {x}")
print(f"Despues de mutar: {mutacion_uniforme(x, 0.5)}")
```

```python
def mutacion_no_uniforme(individuo, probabilidad):
    genes = individuo[0]
    limites = individuo[1]

    for i in range(len(genes)):
        
        x_min = limites[i][0]
        x_max = limites[i][1]

        sigma = ((x_max - x_min) / 6)

        print(sigma)

        p = np.random.random()
        if p < probabilidad:
            valor_muestreado = np.random.normal(scale=sigma)
            gen_nuevo = genes[i] + valor_muestreado

            if gen_nuevo > x_max:
                genes[i] = x_max
            elif gen_nuevo < x_min:
                genes[i] = x_min
            else:
                genes[i] = gen_nuevo
    
    individuo[0] = genes

    return individuo
```

```python
def sigmas_t(current_sigma: list):
    t: float = 1 / np.sqrt(len(current_sigma))
    for i in range(len(current_sigma)):
        nueva_sigma = current_sigma[i] * math.exp(t * np.random.normal())
        current_sigma[i] = nueva_sigma

        if current_sigma[i] < 0.1:
            current_sigma[i] = 0.1 
    
    return current_sigma
```

```python
def mutacion_adaptativa(individuo, probabilidad, sigma):

    for i in range(len(individuo)):
        print(sigma)
        p = np.random.random()
        if p < probabilidad:
            gen_nuevo = individuo[i] + (sigma[i] * np.random.normal())
            individuo[i] = gen_nuevo
        sigma = sigmas_t(sigma)

    return individuo
```
